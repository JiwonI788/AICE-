```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target #iris Îç∞Ïù¥ÌÑ∞Ïùò Ï†ïÎãµÍ∞í 'target' ÎèÑ Í∞ôÏùÄ arrayÏóê Ïó¥ÏúºÎ°ú Ï∂îÍ∞ÄÌï¥Ï§å.
df.head()#Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏. ÏïûÏùò 5Ï§ÑÎßå Î≥¥Ïó¨Ï§å.
```

```python
# Í≤∞Ï∏°Ïπò ÌôïÏù∏ : Î®∏Ïã†Îü¨ÎãùÏóêÏÑú Îç∞Ïù¥ÌÑ∞Ïóê nullÏù¥ ÏûàÏúºÎ©¥ Ïã§ÌñâÏóê Ïò§Î•òÍ∞Ä ÎÇòÍ∏∞ÎïåÎ¨∏Ïóê Íº≠ ÌôïÏù∏ÏùÑ Ìï¥Ï§òÏïºÌï®!
df.isnull().sum()
```

```python
# seaborn ÏùÑ Ïù¥Ïö©Ìïú Í∑∏ÎûòÌîÑ ÏãúÍ∞ÅÌôî
import matplotlib.pyplot as plt#seabornÏúºÎ°ú Í∑∏Î¶∞ Í∑∏ÎûòÌîÑÎ•º Î≥¥Ïó¨Ï£ºÍ∏∞ ÏúÑÌïú Ïó≠Ìï†.
import seaborn as sns#1.Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôîÎ•º ÏâΩÍ≤å Ìï¥Ï£ºÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨. 2.ÌÜµÍ≥Ñ Í∑∏ÎûòÌîÑÏóê Ï†ÅÌï©. 3.matplotlib Î≥¥Îã§ ÏòàÏÅòÍ≥† Í∞ÑÎã®ÌïòÍ≤å Í∑∏Î¶¥ Ïàò ÏûàÏùå.

sns.pairplot(df, hue='target')#ÏÇ∞Ï†êÎèÑÎ•º Í∑∏Î†§Ï£ºÎäî Ìï®Ïàò. targetÏóê Îî∞Îùº Îã§Î•∏ ÏÉâÏúºÎ°ú ÌëúÏãú.
plt.title("Iris data visualiaztion")
plt.show()
```
